```python
import pandas as pd
```

```python
df = pd.read_csv("census.csv")
```

```python
df.info()
```

```python
df.sample(10)
```

```python
# Renamed a column and dropped the columns that I was not interested on.
df = df.rename(columns={"data item": "description"})
df.drop(["state", "state fips"], axis=1, inplace=True)
```

```python
df
```

```python
# Used the melt method to unpivot the table. Those three coulumns seemed to make the most sense
# to use as the variables.

melted_df = df.melt(id_vars=["commodity","description", "domain category"],
                   var_name = "year",
                   value_name = "us_totals")
melted_df
```

```python
melted_df.head(50)
```

```python
# Moved Year to the begining of the index.
first_column = melted_df.pop('year')
melted_df.insert(0, 'year', first_column)

melted_df
```

```python
dc = melted_df.sort_values(by=["domain category"], ascending=True)
```

```python
dc.head(60)
```

```python
#Removed "farm operations" since we dont want it and "commodity totals"
```

```python
df = dc[dc['commodity'].str.contains("FARM OPERATIONS")==False]
df = df[df['commodity'].str.contains("COMMODITY TOTALS")==False]
df
```

```python
df.sample(60)
```

```python
#Discarded any totals in order to refine the data further. 

discard = ["TOTALS"]
df = df[~df.commodity.str.contains('|'.join(discard))]
df
```

```python
# removing "domain category" and sorting by "year".

#This is the final clean data set.

df.drop(["domain category"], axis=1, inplace=True)

df = df.sort_values(by=["year"], ascending=True)
df
```
